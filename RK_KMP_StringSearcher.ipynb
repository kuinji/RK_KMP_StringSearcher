```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
colNames = ['strLen', 'textLen', 'timeKMP', 'timeRK']
dataExp1 = pd.read_csv('resource\\exp1.csv', header = None, names = colNames)
dataExp2 = pd.read_csv('resource\\exp2.csv', header = None, names = colNames)
dataExp3 = pd.read_csv('resource\\exp3.csv', header = None, names = colNames)
```

```python
dataExp1.head()
```

```python
fig = plt.figure(figsize=(7,7))
ax = fig.add_subplot(projection='3d')
ax.plot(dataExp1.textLen, dataExp1.strLen, dataExp1.timeRK, label = 'RK', linewidth = 2, color = 'orange')
ax.set_xlabel('text length')
ax.set_ylabel('substring length')
ax.set_zlabel('T, sec')
ax.legend()
```

```python
fig = plt.figure(figsize=(7,7))
ax = fig.add_subplot(projection='3d')
ax.plot(dataExp1.textLen, dataExp1.strLen, dataExp1.timeKMP, label = 'KMP', linewidth = 2)
#ax.plot(dataExp1.textLen, dataExp1.strLen, dataExp1.timeRK, label = 'RK', linewidth = 2)
ax.set_xlabel('text length')
ax.set_ylabel('substring length')
ax.set_zlabel('T, sec')
ax.legend()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(dataExp2.strLen[::5], dataExp2.timeKMP[::5], label = 'KMP')
plt.plot(dataExp2.strLen[::5], dataExp2.timeRK[::5], label = 'RK')
plt.title('Y=(a)^m; X=({a, b})^(10^6+1); m = 1,...,10^6+1')
plt.xlabel('substring length')
plt.ylabel('T, sec')
plt.legend()
plt.show
```
